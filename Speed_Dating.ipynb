# Cleaning

After looking at the documentation provided by tinder, I decided to see if the similarities or attributes of each participant influence their decisions.

We're going to have to do some selection and cleaning in the data.

I kept the columns describing the participants (age, origin, education, work, income, etc.), their centers of interest, their preference criteria and the scorecards. I removed waves 6 to 9 because they didn't have the same attribute scoring system.

```python
dataset = speed_dating_data[['iid', 'wave', 'age', 'gender_2', 'match', 'dec', 'dec_o', 'attr_o',  'attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'like', 'attr1_1', 'sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1',
                            'samerace', 'age_o', 'race_o','pf_o_att', 'int_corr', 'field_cd', 'mn_sat', 'tuition', 'race', 'income', 'date', 'go_out', 'career_c', 'sports', 'tvsports', 'exercise', 'dining',
                            'museums', 'art', 'hiking', 'gaming', 'clubbing', 'reading', 'tv', 'theater', 'movies', 'concerts', 'music', 'shopping', 'yoga']]

# remove waves 6 to 9
dataset = dataset[~dataset['wave'].isin([6, 7, 8, 9])]
```

```python
print(f"Dataset shape: {dataset.shape}")
print()

print("Percentage of missing values: ")
display(100*dataset.isnull().sum()/dataset.shape[0])
```

```python
# drop columns
drop_columns = ['mn_sat', 'tuition', 'income']
dataset.drop(drop_columns, axis=1, inplace = True)

#drop missing values on all columns with 1.08%
dataset = dataset.dropna(subset=['sports'])

# Filling missing values with mean
means = dataset.mean()
dataset = dataset.fillna(means)

# Final dataset
print(dataset.shape)
print()
print("Percentage of missing values: ")
display(100*dataset.isnull().sum()/dataset.shape[0])
```

```python
dataset.head()
```

```python
# balanced dataset
dataset['gender_2'].value_counts()
```

For missing values, I removed columns with more than 40% missing values. We could also see that there were many columns with 1.08% missing values. This is the questionnaire that participants were asked to fill in beforehand. Some had not filled it in. I decided to remove all these rows, otherwise they would only be filled in with averages, which would strongly bias the data and give us no real additional information. I replaced the rest with averages.

Now we have a dataset that's a little cleaner, still balanced between men and women, with usable columns and what interests us.

# Match and decision

To have a match, both people must have checked "yes" in "decision".

```python
# percentage of match
print("Match percentage: ")
proportion = round((dataset['match'].sum() / len(dataset['match'])) * 100, 2)
print(f"There is {proportion} % of match.")
print()

# percentage of decision
print("Decision percentage: ")
proportion = round((dataset['dec'].sum() / len(dataset['dec'])) * 100, 2)
print(f"There is {proportion} % of match.")
print()

#visualisation
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Barplot Match
match_counts = dataset['match'].value_counts()
sns.barplot(x=match_counts.index, y=match_counts.values,  ax=axes[0])
axes[0].set_title('Match')

# Barplot Decision
dec_counts = dataset['dec'].value_counts()
sns.barplot(x=dec_counts.index, y=dec_counts.values, ax=axes[1])
axes[1].set_title('Decision')

plt.tight_layout()
plt.show()
```

```python
# Tendency to agree to meet again

fig = px.histogram(data_frame=dataset, x='dec', color='gender_2', barmode='group', title='Agree to meet again')
fig.update_layout(bargap=0.2)

fig.show()
```

We can see that 42% of the time a person agrees to see the speed dating partner again, but that there are only 16% of matches, which is rather low. Here we notice that men are more likely to say yes than women.\

I wanted to see if having things in common would increase the chances of a match, so I selected categories such as interests, origin, lifestyle and work.

```python
#let's see if common point are factors for decision and match
common_point = dataset[['match', 'dec', 'samerace', 'age_o', 'race_o','pf_o_att', 'int_corr', 'field_cd', 'race', 'date', 'go_out',
                     'career_c', 'sports', 'tvsports', 'exercise', 'dining', 'museums', 'art', 'hiking', 'gaming', 'clubbing', 'reading', 'tv',
                     'theater', 'movies', 'concerts', 'music', 'shopping', 'yoga']]
```

```python
# Correlations with match and decision variable
corr_match_decision = (common_point.corr()).loc[['match', 'dec']]
f, ax = plt.subplots(figsize=(14, 2))
sns.heatmap(corr_match_decision, ax=ax);
```

But we can see quite clearly that none of these points has much impact on the choices made by the participants.

So I turned to people's characteristics, looking first at what they were most attracted to.

```python
# attributes searched by gender
attributes_factors = dataset[['attr1_1', 'sinc1_1','intel1_1','fun1_1','amb1_1','shar1_1', 'gender_2']].groupby('gender_2').mean()
attributes_factors.columns = ['Attractive', 'Sincere', 'Intelligent', 'Fun', 'Ambitious', 'Shared_Interests']
attributes_factors
```

We can see that men place far more importance on appearance than women, and very little on ambition.
Women, on the other hand, are a little more balanced in their desires, and place primary importance on men's intelligence.
In fact, we can see that shared interests are rarely solicited, which confirms the graph above.

```python
# Correlation
scorecard = dataset[['dec', 'attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'like']]

corr_scorecard = (scorecard.corr()).loc[["dec"]]
f, ax = plt.subplots(figsize=(14, 2))
sns.heatmap(corr_scorecard, annot=True, ax=ax);
```

```python
# Dataset for woman data
woman_data = dataset[(dataset['gender_2'] == 'Woman')]
woman_data.shape
```

```python
# Correlation for woman
scorecard_woman = woman_data[['match', 'dec', 'dec_o', 'attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'like']]

corr_scorecard_woman = (scorecard_woman.corr()).loc[["match", "dec"]]
f, ax = plt.subplots(figsize=(14, 2))
sns.heatmap(corr_scorecard_woman, annot=True, ax=ax);
```

```python
# dataset with man data
man_data = dataset[(dataset['gender_2'] == 'Man')]
man_data.shape
```

```python
# Correlation for man
scorecard_man = man_data[['match', 'dec', 'dec_o', 'attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'like']]

corr_scorecard_man = (scorecard_man.corr()).loc[["match", "dec"]]
f, ax = plt.subplots(figsize=(14, 2))
sns.heatmap(corr_scorecard_man, annot=True, ax=ax);
```

Generally speaking, we can see that attractiveness has a strong influence on people's decisions, but also, and above all, on their general appreciation of a person. Here, however, we can see that a person's fun and shared interests are strongly correlated with a person's decision.

For women, we can finally see that attractiveness is more strongly correlated with their decision than intelligence. As do fun and general appreciation.

For men, attractiveness is a very important factor in their choice. We can also see that their match is also highly dependent on the woman's decision, which can be explained by the fact that, as we saw above, woman say yes less often.

# Conclusion

 - For everyone, attractiveness, shared interests and fun are more important than sincerity, intelligence or ambition.
 - Men attach more importance to their partner's physical appearance, they know it and they assume it. Women attach almost as much importance without being aware of it.
 - The women who take part tend to be younger than the men and more likely to refuse to see the person again.
